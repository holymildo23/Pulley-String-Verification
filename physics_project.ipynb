```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Experimental Data
# ==========================

theta_exp = np.array([0, 0.5, 1.0, 1.5, 2.0, 2.5, np.pi])
T_exp = np.array([10.0, 11.8, 13.2, 16.0, 18.5, 21.0, 25.8])

# ==========================
# Theory
# ==========================

T1 = 10      # Initial tension (N)
mu = 0.3     # Friction coefficient

theta = np.linspace(0, np.pi, 500)

T_theory = T1 * np.exp(mu * theta)

# ==========================
# Plot
# ==========================

plt.figure(figsize=(8,5))

plt.plot(theta,
         T_theory,
         linewidth=2,
         label='Capstan Equation')

plt.scatter(theta_exp,
            T_exp,
            s=60,
            label='Experimental Data')

plt.xlabel(r'Wrapping Angle $\theta$ (rad)')
plt.ylabel(r'Tension $T(\theta)$ (N)')
plt.title('Experimental vs Theoretical Tension Distribution')

plt.legend()
plt.grid(True)

plt.savefig('figure1_tension_distribution.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

mu = np.linspace(0,1,100)

m1 = np.linspace(0.5,5,100)

MU, M1 = np.meshgrid(mu,m1)

M = 4*M1*(np.exp(np.pi*MU)-1)

fig = plt.figure(figsize=(9,6))

ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(MU,
                       M1,
                       M,
                       cmap='viridis')

ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$m_1$ (kg)')
ax.set_zlabel(r'$M$ (kg)')

ax.set_title('Non-Slipping Condition')

fig.colorbar(surf, shrink=0.6)

plt.savefig('figure2_non_slipping.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```
